6. Melakukan Data Cleansing

Kode ini memastikan dataset bersih, konsisten, dan siap dianalisis dengan cara:

a. Menghapus duplikasi,

b. Menstandarkan nama kolom dan tipe data,

c. Mengisi data kosong dengan nilai yang sesuai,Serta memperbaiki atau membuat NIM yang hilang secara otomatis.

Hasil akhirnya adalah dataset yang bebas duplikasi, tidak memiliki nilai kosong kritis, dan memiliki identitas mahasiswa yang lengkap serta valid.

```python
mahasiswa.drop_duplicates(inplace=True)
minat.drop_duplicates(inplace=True)

# Bersihkan nama kolom dan tipe data
mahasiswa.columns = mahasiswa.columns.map(str).str.strip().str.lower()
minat.columns = minat.columns.map(str).str.strip().str.lower()

mahasiswa['nim'] = mahasiswa['nim'].astype(str)
minat['nim'] = minat['nim'].astype(str)

# Ubah nilai kosong menjadi 'Tidak Minat' atau 'Minat'
for df in [minat]:
    for col in df.columns:
        if col not in ['nim', 'nama']:
            df[col] = df[col].apply(
                lambda x: 'Tidak Minat' if pd.isna(x) or str(x).strip() == '' else 'Minat'
            )

# Lengkapi NIM kosong berdasarkan nama yang sama
nama_nim_map = mahasiswa.loc[mahasiswa['nim'] != '', ['nama', 'nim']].drop_duplicates().set_index('nama')['nim'].to_dict()

def isi_nim(row, existing_nims):
    if row['nim'] == '' or pd.isna(row['nim']):
        nama = row['nama']
        if nama in nama_nim_map:
            return nama_nim_map[nama]  # gunakan nim dari nama yang sama
        else:
            # buat nim unik baru (format: 999001, 999002, dst)
            new_nim = f"999{len(existing_nims)+1:03d}"
            existing_nims.add(new_nim)
            return new_nim
    return row['nim']

# Gunakan set untuk melacak nim unik baru
nim_terpakai = set(mahasiswa['nim'].dropna().unique())
mahasiswa['nim'] = mahasiswa.apply(lambda r: isi_nim(r, nim_terpakai), axis=1)

nim_terpakai = set(minat['nim'].dropna().unique())
minat['nim'] = minat.apply(lambda r: isi_nim(r, nim_terpakai), axis=1)

print("\n Proses cleansing selesai dengan aman dan kolom status_minat berhasil dibuat.")
print(mahasiswa.head())
```

7. Mencocokan Data

Kode ini berfungsi untuk memastikan kedua dataset:

a. Memiliki format data yang seragam

b. Tidak mengandung data bermasalah

c. Memiliki NIM serta nama yang benar-benar cocok sebelum dilakukan proses merge.

Dengan ini hasil penggabungan data nantinya akan akurat, bersih, dan bebas dari ketidaksesuaian.

```python
for df in [mahasiswa, minat]:
    # Pastikan NIM berupa string dan hanya berisi angka
    df['nim'] = df['nim'].astype(str)
    df['nim'] = df['nim'].str.replace(r'\D', '', regex=True)
    df['nim'] = df['nim'].str.strip().str.lower()

    # Normalisasi nama jadi huruf kecil dan spasi tunggal
    df['nama'] = df['nama'].astype(str)
    df['nama'] = df['nama'].str.replace(r'\s+', ' ', regex=True).str.strip().str.lower()

# Hapus nama yang bermasalah
target_nama = "aris tarihoran"
mahasiswa = mahasiswa[mahasiswa['nama'] != target_nama]
minat = minat[minat['nama'] != target_nama]

# Cek kesesuaian NIM dan Nama sebelum merge
nim_diff = set(mahasiswa['nim']) - set(minat['nim'])
nama_diff = set(mahasiswa['nama']) - set(minat['nama'])

if nim_diff or nama_diff:
    print("\n Ada data yang belum cocok antara dua dataset:")
    print(f" Jumlah NIM tidak cocok: {len(nim_diff)}")
    print(f" Jumlah Nama tidak cocok: {len(nama_diff)}")

    max_len = max(len(nim_diff), len(nama_diff))
    diff_report = pd.DataFrame({
        "nim_tidak_cocok": list(nim_diff)[:max_len] + [""] * (max_len - len(nim_diff)),
        "nama_tidak_cocok": list(nama_diff)[:max_len] + [""] * (max_len - len(nama_diff))
    })

    print("\nContoh data yang tidak cocok:")
    display(diff_report.head(10))
else:
    print("\n Semua NIM dan Nama sudah cocok, siap digabung!")
```

8. Gabungkan Dataset Dan mendowload Hasil EDA

Kode tersebut berfungsi untuk menggabungkan dua dataset, yaitu data mahasiswa dan data minat mahasiswa, menggunakan metode inner join berdasarkan kolom nim dan nama. Hasil penggabungan ditampilkan untuk memastikan data telah tergabung dengan benar, kemudian disimpan sebagai file "Hasil EDA Data Mahasiswa.csv" menggunakan to_csv(). Selanjutnya, perintah files.download() digunakan agar file hasil merge dapat langsung diunduh dari Google Colab. Secara keseluruhan, kode ini memudahkan proses penggabungan dan penyimpanan data sehingga siap digunakan untuk analisis lebih lanjut.

```python
merged = pd.merge(mahasiswa, minat, on=['nim', 'nama'], how='inner')
# merged = pd.merge(mahasiswa, minat, on='nim', how='inner')
print("\n=== Data Setelah Merge ===")
display(merged.head())
print("Jumlah data setelah merge:", merged.shape)

# Simpan hasil merge ke file CSV
output_path = "/content/Hasil EDA Data Mahasiswa.csv"
merged.to_csv(output_path, index=False)
print(f"\n Dataset hasil merge berhasil disimpan di: {output_path}")

# Buat link download langsung (hanya di Google Colab)
from google.colab import files
files.download(output_path)
```

9. Statistik Deskriptif

Kode ini digunakan untuk:

a. Menghasilkan ringkasan statistik dari seluruh kolom dalam dataset,

b. Menampilkan distribusi nilai pada kolom bertipe kategorikal,

c. Membantu memahami struktur serta karakteristik umum data, seperti sebaran nilai numerik maupun dominasi kategori tertentu.

Dengan ini bisa dengan cepat mengenali kondisi dan pola dasar dari data hasil penggabungan.

```python
print("\n--- Statistik Deskriptif ---")

# Membulatkan hasil describe ke 2 angka di belakang koma
stat_desc = merged.describe(include='all').round(2)
display(stat_desc)

# Menampilkan frekuensi untuk kolom kategorikal
categorical_cols = merged.select_dtypes(include='object').columns
for col in categorical_cols:
    print(f"\nFrekuensi untuk {col}:")
    print(merged[col].value_counts().head())
```

10. Visualisasi Data

Kode ini berfungsi untuk:

a. Memvisualisasikan pola distribusi data numerik dan kategorikal

b. Menunjukkan proporsi atau dominasi kategori tertentu

c. Menganalisis hubungan antar variabel penting (contohnya IPK dan semester).

Dengan visualisasi ini bisa memahami karakteristik data secara menyeluruh.

```python
numerical_cols = merged.select_dtypes(include=np.number).columns
categorical_cols = merged.select_dtypes(include='object').columns

# Distribusi untuk kolom numerik
for col in numerical_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(merged[col], kde=True, color='skyblue')
    plt.title(f'Distribusi {col}')
    plt.tight_layout()
    plt.show()

# Distribusi untuk kolom kategorikal (yang uniknya < 10)
for col in categorical_cols:
    if merged[col].nunique() < 10:
        plt.figure(figsize=(6,4))
        sns.countplot(y=merged[col], palette="muted")
        plt.title(f'Jumlah Data Berdasarkan {col}')
        plt.tight_layout()
        plt.show()

# Korelasi spesifik (contoh: IPK vs Semester)
if {'ipk', 'semester'}.issubset(merged.columns):
    plt.figure(figsize=(6,4))
    sns.scatterplot(x='semester', y='ipk', data=merged, hue='jenis_kelamin', palette='Set2')
    plt.title('Korelasi antara Semester dan IPK')
    plt.tight_layout()
    plt.show()
```

11. Heatmap Korelasi

Kode tersebut digunakan untuk menampilkan peta korelasi (heatmap) antar variabel numerik dalam dataset merged, dengan tujuan memahami hubungan antar angka seperti IPK, umur, semester, atau nilai lainnya.

```python
if len(numerical_cols) > 1:
    print("\n--- Heatmap Korelasi Antar Variabel Numerik ---")
    plt.figure(figsize=(10,8))
    sns.heatmap(
        merged[numerical_cols].corr().round(2),
        annot=True,
        cmap='coolwarm',
        fmt=".2f",
        linewidths=0.5
    )
    plt.title('Heatmap Korelasi Antar Variabel Numerik')
    plt.tight_layout()
    plt.show()
```
